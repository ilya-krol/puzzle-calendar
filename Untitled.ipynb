```python
import puzzcal as pc

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools
from colour import Color
```

```python
solutions = pd.DataFrame(pc.correct_solutions(), columns =['month', 'day', 'solution'])
sl = solutions
```

```python
groupByMonthDay = solutions.groupby(['month','day'])['solution'].count()
```

```python
groupByMonthDay.plot(kind = 'area',figsize=(20,10))
plt.savefig('groupByMonthDay.png')
plt.show()
```

```python
groupByMonthDay.plot(kind = 'hist',figsize=(20,10), bins=38, xlim=(0,220), xticks = np.arange(7, 220, 11) )
plt.savefig('groupByMonthDayHist.png')
plt.show()
```

```python
groupByMonthDay.plot(kind = 'box',figsize=(20,5), vert=False )
plt.savefig('groupByMonthDayBox.png')
plt.show()
```

```python
red = Color("#ff7000")
yellow = Color("#eaf800")
green = Color("#79cb00")
colors1 = list(red.range_to(yellow, 54))
colors2 = list(yellow.range_to(green, 48))
colors = colors1 + colors2[1:]
```

```python
monthMin = groupByMonth.min()
monthMax = groupByMonth.max()
mons = pd.DataFrame(groupByMonth)
mons['complex'] = (mons['solution']-monthMin) / (monthMax-monthMin)
mons['color'] = mons['complex'].transform( lambda x:  colors[int(x*100)])


mons
```

```python
dayMin = groupByDay.min()
dayMax = groupByDay.max()

days = pd.DataFrame(groupByDay)
days['complex'] = (days['solution']-dayMin) / (dayMax-dayMin)
days['color'] = days['complex'].transform( lambda x:  colors[int(x*100)])

days
```
